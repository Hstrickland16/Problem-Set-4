## Visualizing SARS-CoV-2 Cases Over Time with the COVID-19 API
This code receieves data from the COVID-19 API regarding confirmed SARS-CoV-2 Cases in Mexico. Using this data, a graph can be displayed depicting the number of COVID-19 cases over time.

Made by: Adithi Jeevan

```python
import requests
import json
import matplotlib.pyplot as plt

#receives data from the API regarding confirmed COVID-19 cases in Mexico
response = requests.get("https://covid-api.mmediagroup.fr/v1/history?ab=MX&status=confirmed")
response.text

#reads the data from the API
try:
    js = json.loads(response.text)
except:
    js = None
    
#returns a sorted list of tuples for dates and confirmed cases on these dates
lists = sorted(js["All"]["dates"].items())

#creates a graph that displays COVID-19 cases over time
x, y = zip(*lists) 
plt.title("Number of COVID-19 Cases in Mexico")
plt.xlabel('Dates')
plt.ylabel('Number of Cases')

plt.plot(x, y)
plt.show()
```
